### I webscrape to create my original dataset consist of rows of Philippines' news headlines 

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

News = []
for i in range(1,4346):
    url = f'https://www.manilatimes.net/business/page/{i}'
    response = requests.get(url).text
    soup = BeautifulSoup(response, 'lxml')
    articles = soup.find_all('div', class_= 'col-1 mr-b20')

    for article in articles:
        article_title = article.find('div', class_ = 'article-title-h4 ellipsis-2 mr-b10').text
        article_date = article.find('span', class_ = 'roboto-a').text
        News.append([article_date, article_title])

display(News)
```

```python
df = pd.DataFrame(News, columns=['Date', 'Headlines'])

# Remove unwanted characters in multiple columns using applymap()
# df = df.applymap(lambda x: x.replace('\n', ''))
df['Headlines'] = df['Headlines'].str.replace('\n', '')
df['Date'] = df['Date'].str.replace('\n-','')
df['Date'] = df['Date'].str.replace('\n','')

display(df)
```

```python
df['Date'] = df['Date'].str.replace(' -','')
```

```python
# Export the dataframe as csv file
df.to_csv('PH-News-MT.csv')
```
